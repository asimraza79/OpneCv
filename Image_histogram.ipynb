# Image Histogram

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

path= r"""C:\Users\RANA ASIM\Pictures\Camera Roll"""

img_name= "color.jpg"

img= cv2.imread(path + '\\' + img_name)

image= cv2.resize(img,(1280,720))

cv2.imshow('Colour Image', image)

cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
image.shape
```

```python
image.ravel().shape
```

```python
plt.hist(image.ravel(), bins=256, range=[0,255])
plt.show()
```

```python
colors=('b', 'g' ,'r')

img_list= [image[:, :, 0].ravel(), image[:, :, 1].ravel(), image[:, :, 2].ravel()]

plt.hist(img_list, color=colors, label=colors,bins=256, range=[0,255])
plt.legend()
plt.show()
```

```python
colors=('b', 'g' ,'r')

img_list= [image[:, :, 0].ravel(), image[:, :, 1].ravel(), image[:, :, 2].ravel()]

plt.hist(img_list, color=colors, label=colors)
plt.legend()
plt.show()
```

```python
colors=('b', 'g' ,'r')

for i, color in enumerate(colors):
    cal_hist = cv2.calcHist([image], [i], None, [256], [0,256])
    plt.plot(cal_hist, color=color)
    
plt.show()
```
