# Save Image from Video

```python
path1=r'C:\Users\RANA ASIM\Videos\My videos\Asim (4).mp4'

os.mkdir('video_to_img')   #Create directory

cap= cv2.VideoCapture(path1)  #Capture video

img_count=1

while cap.isOpened():
    ret, frame= cap.read()   #read video frame
    
    if not ret:
        print('Unable to read frame')
        break
        
    is_img_write=cv2.imwrite(f'video_to_img\image{img_count}.jpeg', frame)
    
    if is_img_write:
        print(f'image save at video_to_img\image{img_count}.jpeg')
        
    cv2.imshow('video',frame)
    
    cv2.waitKey(25) & 0xff == ord('q')
    img_count += 1
    
cap.release()
cv2.destroyAllWindows()
```
